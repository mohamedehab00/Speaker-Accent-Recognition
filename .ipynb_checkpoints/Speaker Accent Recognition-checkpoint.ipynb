```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("accent-mfcc-data-1.csv")
```

```python
df_copy = df.copy()
```

```python
df.head()
```

```python
langs_info = df["language"].value_counts()
print(langs_info)
```

```python
plt.bar(langs_info.index,langs_info)
```

```python
labelencoder = LabelEncoder()
df['language'] = labelencoder.fit_transform(df['language'])
```

```python
df.info()
```

```python
x,y = plt.subplots(figsize=(12,9))
sns.heatmap(df.corr(),cmap='YlGnBu' , square=True , linewidth= .5, annot = True)
```

```python
df.shape
```

```python
np.random.seed(77)
ids = np.random.permutation(len(df))
precent = 0.8
train_limit = round(precent * len(df))
print(train_limit)
train_ids = ids[:train_limit+1]
test_ids = ids[train_limit+1:]
print(train_ids)
print(test_ids)
x_train_data = df.iloc[train_ids,0:-1]
x_test_data = df.iloc[test_ids,0:-1]
y_train_data = df.iloc[train_ids,-1:]
y_test_data = df.iloc[test_ids,-1:]
```

```python
x_train_data
```

```python
y_train_data
```

```python
clf = DecisionTreeClassifier()
clf.fit(x_train_data, y_train_data)
pred = clf.predict(x_test_data)
```

```python
print(y_test_data)
```

```python
print('Prediction Accuracy : {}%'.format(round(accuracy_score(pred,y_test_data)*100)))#prediction accuracy
```

```python
knn_model = KNeighborsRegressor(n_neighbors=1)
knn_model.fit(x_train_data, y_train_data)
train_preds = knn_model.predict(x_test_data)
print(y_test_data.head())
train_preds = np.ceil(train_preds)
print(train_preds.reshape(65))
train_preds.shape
print('Prediction Accuracy : {}%'.format(round(accuracy_score(train_preds,y_test_data)*100)))#prediction accuracy
```

```python
df_copy.head()
```

```python
def modify(x):
    if x != 'US':
        return 'non-US'
    else:
        return x
```

```python
df_copy['language'] = df_copy['language'].apply(modify)
```

```python
df_copy.head()
```

```python
langs_info_copy = df_copy["language"].value_counts()
plt.bar(langs_info_copy.index,langs_info_copy)
```

```python
df_copy['language'] = labelencoder.fit_transform(df_copy['language'])
```

```python
df_copy["language"].value_counts()
```

```python
np.random.seed(222)
ids = np.random.permutation(len(df_copy))
precent = 0.8
train_limit = round(precent * len(df_copy))
train_ids = ids[:train_limit+1]
test_ids = ids[train_limit+1:]
print(train_ids)
print(test_ids)
x_train_data_copy = df_copy.iloc[train_ids,0:-1]
x_test_data_copy = df_copy.iloc[test_ids,0:-1]
y_train_data_copy = df_copy.iloc[train_ids,-1:]
y_test_data_copy = df_copy.iloc[test_ids,-1:]
```

```python
x_train_data_copy
```

```python
y_train_data_copy
```

```python
knn_model_copy = KNeighborsRegressor(n_neighbors=1)
knn_model_copy.fit(x_train_data_copy, y_train_data_copy)
train_preds_copy = knn_model_copy.predict(x_test_data_copy)
train_preds_copy.reshape(train_preds_copy.shape[0])
print('Prediction Accuracy : {} %'.format(round(accuracy_score(train_preds_copy,y_test_data_copy)*100)))#prediction accuracy
```

```python
# Split dataset into training set and test set
feature_cols=['X1','X2','X3','X4','X5','X6','X7','X8','X9','X10','X11','X12']
x = df_copy[feature_cols]
y = df_copy.language
x_train_data_copy, x_test_data_copy, y_train_data_copy, y_test_data_copy = train_test_split(x, y, test_size=0.1, random_state=100) # 90% training and 10% test
```

```python
knn_model_copy.fit(x_train_data_copy, y_train_data_copy)
train_preds_copy = knn_model_copy.predict(x_test_data_copy)
train_preds_copy.reshape(train_preds_copy.shape[0])
print('Prediction Accuracy : {} %'.format(round(accuracy_score(train_preds_copy,y_test_data_copy)*100)))#prediction accuracy
```

```python
import librosa
import librosa.display
import IPython.display as ipd
audio = "US-F-4-1.wav"
```

```python
ipd.Audio(audio)
```

```python
signal, sr = librosa.load(audio,sr=44100)
```

```python
signal
```

```python
sr
```

```python
mfccs = librosa.feature.mfcc(y=signal, n_mfcc=13, sr=sr)
print(mfccs)
print(mfccs[0].min())
print(mfccs[0].max())
print(mfccs.shape)
```

```python
mean_vec = []

for i in mfccs:
    mean_vec.append(i.mean())
    
mean_vec =  np.array(mean_vec)
mean_vec = mean_vec[:-1]
mean_vec
```

```python
p = knn_model_copy.predict(mean_vec.reshape(1, -1))
print(mean_vec.reshape(1, -1))
print(p)
```
